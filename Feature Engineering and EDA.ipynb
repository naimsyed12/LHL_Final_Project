```python
##Feature Engineering: Creating a feature that indicates if a player got drafted using boolean 
df['Drafted'] = (df['pick'] > 0).astype(int)
```

```python
##Value Counts show this is an imbalanced dataset providing more undrafted (0) players than drafted (1)
df['Drafted'].value_counts()
```

```python
##Stats for drafted players
df[['pts','ast','treb','stl','blk']].groupby(df['Drafted']).median()
```

## Data Cleaning

```python
df_encoded = pd.get_dummies(df, columns=['Position'], prefix='Position',dtype=int)
```

```python
df_encoded.head()
```

```python
df_encoded['year']=df_encoded['year'].astype(int)
```

```python
df_encoded.info()
```

```python
##replacing null picks (undrafted) with 61 (last pick +1)
df_encoded['pick']=df_encoded['pick'].fillna(61)
```

```python
df_encoded['pick'].value_counts()
```

```python
## Over 13 years, maximum count for each pick position should only be 13. More than 13 is an indicator for duplicates. 
df['pick'].value_counts()
```

```python
## Remove categorical that has not be encoded and group stats by player id (pid). Using max agg function to keep best player season.
single_season_df = df_encoded.groupby(['pid'])[['GP', 'Min_per', 'Ortg', 'usg', 'eFG',
       'TS_per', 'ORB_per', 'DRB_per', 'AST_per', 'TO_per', 'FTM', 'FTA',
       'FT_per', 'twoPM', 'twoPA', 'twoP_per', 'TPM', 'TPA', 'TP_per',
       'blk_per', 'stl_per', 'ftr', 'porpag', 'adjoe', 'pfr', 'year',
       'pid', 'Rec Rank', 'ast/tov', 'pick', 'drtg', 'adrtg',
       'dporpag', 'stops', 'bpm', 'obpm', 'dbpm', 'gbpm', 'mp', 'ogbpm',
       'dgbpm', 'oreb', 'dreb', 'treb', 'ast', 'stl', 'blk', 'pts', 'Drafted',
       'Position_C', 'Position_Combo G', 'Position_PF/C', 'Position_Pure PG',
       'Position_Scoring PG', 'Position_Stretch 4', 'Position_Wing F',
       'Position_Wing G']].max()
```

```python
###After grouping, maximum count is 13. Pick positions with less than 13 are due to dataset not including european players.
single_season_df['pick'].value_counts()
```

```python
##Checking Nulls
single_season_df.isnull().sum()
```

```python
## Replace null values with medians for each column
for column in single_season_df.columns:
    if single_season_df[column].isnull().any():
        median = single_season_df[column].median()
        single_season_df[column].fillna(median, inplace=True)
```

```python
single_season_df.isnull().sum()
```

## Data Visualizations

```python
##Average Draft Pick Position based on Player Position
plt.figure(figsize=(12,8))
sns.barplot(data=single_season_df,x='year',y='pts',hue='Drafted')
plt.ylabel('Points per Game')
plt.xlabel('Year')
plt.legend(title = 'Drafted', loc='upper right')
plt.title('Points per Game by Draft Status')
plt.show()
```

```python
##Average Draft Pick Position based on Player Position
plt.figure(figsize=(15,6))
sns.scatterplot(data=single_season_df,x='twoPM',y='twoPA',hue='Drafted')
plt.ylabel('twoPA')
plt.xlabel('twoPM')
plt.legend(title = 'Drafted', loc='upper right')
plt.title('Points vs Assists')
plt.show()
```

```python
## Average Points across draft position
plt.figure(figsize=(25, 8))
sns.barplot(data=single_season_df, x='pick', y='pts')
plt.show()
```

```python
##Average Rebounds across draft position
plt.figure(figsize=(25, 8))
sns.barplot(data=single_season_df, x='pick', y='treb')
plt.show()
```

```python
## Stats for First Overall Pick Over Time
plt.figure(figsize=(20,8))
sns.lineplot(data = single_season_df[single_season_df['pick']==1], x = 'year', y = 'pts', label = 'Points')
sns.lineplot(data = single_season_df[single_season_df['pick']==1], x = 'year', y = 'ast', label = 'Assists')
sns.lineplot(data = single_season_df[single_season_df['pick']==1], x = 'year', y = 'treb', label = 'Rebounds')
sns.lineplot(data = single_season_df[single_season_df['pick']==1], x = 'year', y = 'blk', label = 'Blocks')
plt.legend()
plt.title('First Overall Pick Stats Over Time')
plt.ylabel('Stats')
plt.xlabel('Year')
plt.show()
```

```python
#Selected Column Names
columns = ['GP', 'Min_per', 'usg', 'eFG', 'TO_per', 'FT_per',
       'twoP_per', 'TP_per', 'year','stops',
       'mp', 'oreb', 'dreb', 'treb', 'ast', 'stl', 'blk',
       'pts']

# Calculation of number of rows/columns for subplots
num_cols = 3
num_rows = (len(columns) + num_cols - 1) // num_cols  # Match number of rows with columns

# Create subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(30, 4 * num_rows))

# Iterate through each column and plot a histogram
axes = axes.flatten()
for i, col in enumerate(columns):
    ax = axes[i]
    ax.hist(single_season_df[col], bins=15, edgecolor='black')
    ax.set_title(col)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
##Points vs Usage Comparison
plt.figure(figsize=(12,8))
sns.scatterplot(data = single_season_df, x = 'usg', y = 'pts', hue = 'Drafted')
plt.title('Points vs Usage')
plt.xlabel('Usage')
plt.ylabel('Points')
plt.show()
```

### Unlike most steps where outliers are removed/adjusted, in this situation outliers are important to predictions. I.e. When most players are scoring 5-10 points per game but one player scores 25 per game, that player is more likely to be drafted since they would be a top prospect. 
### Points vs usage show how even with higher usage rates, it doesn't result in higher scoring and instead can lower a players chances to getting drafted if they are an inefficient scorer (bottom right). Players closer to the top right are more effective scorers which would increase their draft prospect.
